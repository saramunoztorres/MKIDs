```python
import numpy as np
import os 
import pandas as pd
import glob
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patches as pc
import seaborn as sns

path = os.getcwd()
```

### File general

Creación de un único archivo ordenado temporalmente, del cual se eliminan columnas innecesarias

```python
os.chdir(path+'\Datos\PruebaCryo-1')    
os.getcwd()

# Array con los nombres de los arhivos ordenados por fecha de creación
files = glob.glob1(path+'\Datos\PruebaCryo-1',"*.csv")
files.sort(key=lambda x: os.path.getmtime(x))
files = [str(x) for x in files]

# Concatenación de archivos
print(files[0])
df = pd.read_csv(files[0])

for i in range(1,len(files)):
    file = files[i]
    print(file)
    df_file = pd.read_csv(file)
    df_file = df_file.drop(df_file.index[0:2])   # Eliminar las dos primeras filas
    df = pd.DataFrame(np.concatenate((df.values,df_file.values)),columns=df.columns) 

# Eliminación de columnas irrelevantes
df = df.drop(df.columns[20:], axis='columns')
```
