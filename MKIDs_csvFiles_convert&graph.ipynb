```python
import string
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(path+'\Datos\MKID\Temperatura\\All 1700mk.csv')

df  = df.rename(columns={'# Version 1.00':'Freq (Hz)',
                         'Unnamed: 1':'S21 (dB)',
                         'Unnamed: 2':'S21 (ang)'})
df = df.drop(df.index[0:2])

freq = df.iloc[:,0]
freq = freq.tolist()

for i in range(len(freq)):
    freq[i] = freq[i].replace(';',',')
    freq[i] = freq[i].split(',')

df = pd.DataFrame(freq,columns=['freq','db','phase','none'])
freq = df.drop('none',axis=1).values.tolist()
 
df = pd.DataFrame(freq)
df.to_csv(path+"/Datos/MKID/Temperatura/" + "1700 mk.csv", mode = 'a', index=False)
```
